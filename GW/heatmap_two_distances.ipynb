# Visualizing the difference between the GW distance and the Mahalanobis Distance

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dist = np.load('../datasets/distance_matrices/distance_matrix_raw_data.npy')

M_dist = np.load('../datasets/distance_matrices/Mahalanobis_distance_matrix.npy')
```

```python
dist
```

```python
print(M_dist)
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,4))


sns.heatmap(dist, cmap='viridis', ax=axes[0])

sns.heatmap(M_dist, cmap='viridis', ax=axes[1])

axes[0].set_title('Gromov-Wasserstein distance')
axes[1].set_title('Mahalanobis distance')

plt.show()
```
